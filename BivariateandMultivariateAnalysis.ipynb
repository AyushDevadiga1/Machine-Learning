<h3> Bivariate Analysis  -  Analysis done on the basis of two columns / variables <br>
Multivariate Analysis - Analysis done on more than two variables using many parameters <h3>

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("Titanic-Dataset.csv")
```

```python
insurance = pd.read_csv("insurance.csv")
insurance
```

We will be doing both the types of analysis simulatenousl on the dataset and the variables can be either : <br>
a).Numeric<br>
b).Categorical <br>
Depending on this we can divide into different types and we have differnt methods for each type

### When we have to do analysis on a lot of data and everything need to be stuided we can use 

```python
iris = pd.read_csv ('iris.csv')
iris
```

```python
sns.pairplot(
                data = iris , 
                hue = 'Species'
)
```

### When both the columns are Numbers we can use :

```python
# Scatterplot to plot the variables and check for linear relationship.
sns.scatterplot(
                    x = insurance['age'],
                    y = insurance['charges'],
)
```

### When many of the columns are Numbers , with some being categorical we can use :

```python
# For multivariate analysis we can add more data like hue , shape , etc
sns.scatterplot(
                    x = insurance['age'],
                    y = insurance['charges'],
                    hue = insurance['smoker'],
                    style = insurance['sex'],
                    # Also size could have been included for changing the scale if the dataset supports
)
```

### When one of the column is numerical and other/s is categorical we can use :

```python
titanic
```

```python
# We can use barplot to check the category vs the numerical column and use extra parameters for multivariate analysis
sns.barplot(
                x = titanic['Pclass'] ,
                y = titanic['Fare']
)
```

```python
# Now adding params
sns.barplot(
                x = titanic['Pclass'] ,
                y = titanic['Fare'] ,
                hue = titanic['Sex']
)
```

```python
# We use boxplot when we need to analysis on the continuous data variables and get deeper statistical insights.
# Also here we have used two categories 
sns.boxplot(
                x = titanic['Sex'],
                y = titanic['Age'],
                hue = titanic['Survived']
)
```

### To find probality , type of distribution and then trends we use

```python
# We use the distplot to check the type of distribution , check for underlying patterns and also check for other things.
# Note that the distplot from older version is depricated now and we can use for histograms the displot and for the kde curves the kdeplot.
sns.kdeplot(    
                titanic[titanic['Survived'] == 0]['Age'] 
)
sns.kdeplot(    
                titanic[titanic['Survived'] == 1]['Age'] 
)
```

### Now if we are dealing with one categorical column and other column/s is categorical , then:

```python
# We can use the heatmap ; before that we are creating a caroll diagram to check and we can pass this in the histogram later.
surv_pclass = pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
#   We use the heatmap to see the intensity and the clusters from the categories.
sns.heatmap(
                data  = surv_pclass,
                cmap = 'magma'
                
)
```

```python
# Another method is to see the clustering that is can we uniquely identify the clusters without the labels given 
sns.clustermap(
                    data = surv_pclass
)
```
