```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('loan_data.csv')
df.head(5)
```

```python
df.columns.unique
```

```python
categorical_without_order = df[['person_gender','person_home_ownership','loan_intent','previous_loan_defaults_on_file']]
categorical_with_order = df[['person_education']]
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder
lbe = LabelEncoder()
```

```python
# Here we do not need the LabelEncoder as the labels or the output column is already in the form of integers
# We use the OneHotEncoder for encoding labels with no ordered label values.

ohe = OneHotEncoder(sparse_output=False)

ohe.fit(categorical_without_order)
ohe_encoded = ohe.transform(categorical_without_order)
type(ohe_encoded)
```

```python
df1 = pd.DataFrame(ohe_encoded , columns = ohe.get_feature_names_out(['person_gender','person_home_ownership','loan_intent','previous_loan_defaults_on_file']))
df1
```

```python
# We do drop = 'first' we get k-1 dimension columns 
# We do this cause we need to avoid multi-collinearity between the columns in algorithms like Linear Regression , Logistic Regression , etc.
ohe1 = OneHotEncoder(drop='first',sparse_output=False)
ohe1.fit(categorical_without_order)
ohe_encoded1 = ohe1.transform(categorical_without_order)
type(ohe_encoded1)

df2 = pd.DataFrame(ohe_encoded1 , columns = ohe1.get_feature_names_out(['person_gender','person_home_ownership','loan_intent','previous_loan_defaults_on_file']))
df2
```

```python
oe = OrdinalEncoder( categories=person_education_order )
education_encoded = oe.fit_transform(categorical_with_order)
categorical_with_order,education_encoded
```

### Now we will see how this can be simplified by ColumnTransformer

```python
from sklearn.compose import ColumnTransformer
# The value accepted by the tranformers list is a tuple with the transfromation being took place .
```

```python
#           Genreal Transformer format : [
#                                          ( 'name_of_the_transformation , Transformer_object(params), object_passed ) ,
#                                              remainder = 'passthrough'
#                                           ]
ct = ColumnTransformer(
                        transformers = [
                                         ( 'ohe_nominal' , OneHotEncoder(drop='first') , categorical_without_order ),
                                         (  'ord_edu' , OrdinalEncoder(categories = person_education_order) , categorical_with_order ),
                                         ('pass_num', 'passthrough', 
                                            [col for col in df.columns if col not in (categorical_without_order + categorical_with_order) ])
                        ]
)
```

```python
df_transformed = ct.fit_transform(df)
df_transformed
```

```python
final_df = pd.DataFrame( df_transformed , columns = ct.get_feature_names_out())
```
