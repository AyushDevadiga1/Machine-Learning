```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv('Titanic-dataset.csv')
diabetes_data = pd.read_csv('diabetes.csv')
loan_data = pd.read_csv('loan_data.csv')
```

```python
# To analyze corr matrix for numeric values we can use the Heatmap 
plt.figure(figsize=(10,8))
sns.heatmap(
    data = diabetes_data.corr(),
    annot= True,
    cmap = 'Reds',
    shading = 'auto',

)
```

```python
# Cluster map gives the heirarchial cluster for the dataset .



sns.clustermap(
    data = diabetes_data,
    cmap = 'Blues',
    standard_scale = 1
)
```

```python
# Facet grids - Multi-plot grid for plotting conditional relationships.
facet_grid = sns.FacetGrid(
    data = titanic_data,
    row = 'Survived',
    col = 'Pclass',
    hue = 'Sex',
    margin_titles=True
)
facet_grid.map(sns.scatterplot,'Age','Fare')
```
