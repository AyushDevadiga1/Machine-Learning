```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('breast_cancer_dataset.csv')
df.head()
```

We basically are predicting whether the cells are malignant or benign with the help of logistic regression using all the columns.
Here basically we are given the measurement of the cell to predict whether the cell is cancer cell forming or not . 
We will use Logistic Regression here ; Logistic Regression is used when we have to classify the object into a class which in our case is binary : Malignant or Benign (0 or 1 in simple words).

## Data Cleaning

```python
#   We found that there is one column which is empty , so we gonna drop 
#   Run this once only if run more times will show errors as columns are already deleted ðŸ˜›
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
df.head()
```

### Convering the target column to categorical type column

```python
#   Now to convert diagnosis to 1s and 0s
df['diagnosis'] = [1 if value == 'M' else 0 for value in df['diagnosis']]
```

```python
df.head()
```

```python
df['diagnosis'] = df.diagnosis.astype('category',copy=False)
df["diagnosis"].value_counts().plot(kind="bar")
```

```python
# Divide into target variables and predictors 
y = df.diagnosis
X = df.drop(['diagnosis'],axis=1) #inplace = True makes the effect permanent here we have just stored the temp disabeled value in predictors
```

```python
X
```

##  Scalling

```python
# Standardization of the units so we can have mean =  0 and std = 1 ; Basically centers value around 0
from sklearn.preprocessing import StandardScaler
```

```python
#Instantiate the StandardScaler 
scaler = StandardScaler()
```

```python
# Fit the scaler in the dataset for predicting columns 
X_scaled = scaler.fit_transform(X)
X_scaled
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split ( X_scaled , y , test_size=0.30 , random_state= 42 )
```

# Training the model 

```python
from sklearn.linear_model import LogisticRegression
```

```python
#Instantiate the Logistic Regression model
lorg = LogisticRegression()
```

```python
#Fitting the data 
lorg.fit(X_train,y_train)
```

```python
# Prediction using the test data
y_prediction = lorg.predict(X_test)
```

```python
y_prediction
```

```python
y_test
```

# Model Evaluation

```python
# Evaluation methods from sklearn
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_prediction)
print(f"The accuracy of the model was : {accuracy : .3f} ")
```

The accuracy found was just amazing as the model had a accuracy of almost 98%.

```python
# Basicallt gives the report of Precision , Recall , F1-score and Support 
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_prediction))
```
