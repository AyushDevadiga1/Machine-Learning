```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
# We take values here cause we want to apply numpy functions and we need a numpy array for that.
X = df.iloc[:,0:1].values
y = df.iloc[:,-1].values
```

```python
type(X),type(y)
```

```python
# Now we will do train test split for experementing with our own model.
X_train , X_test , y_train , y_test = train_test_split ( X ,y, test_size=0.2 , random_state=2)
```

```python
X_train.shape[0]
```

```python
class LinearRegression:
    def __init__(self):
        self.m = None
        self.b = None
        self.X_test = None 
    def fit(self,X_train,y_train):
        n = 0
        d = 0
        X_train_mean = np.mean(X_train)
        y_train_mean = np.mean(y_train)

        for x in range(X_train.shape[0]):
            n = n + ( X_train[x] - X_train_mean ) * (y_train[x] - y_train_mean )
            d = d + ( X_train[x] - X_train_mean ) * ( X_train[x] - X_train_mean )

        self.m = n/d
        self.b = y_train_mean - self.m * X_train_mean
        
        print(f"Slope is : {self.m} ")
        print(f"Intercept is : {self.b} ")


    def predict(self,x_test):
        return (self.m * X_test + self.b)

    
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Residual Analysis
residual = y_test - y_pred
residual.sum(),residual.mean()
```
