```python
# Now as we saw earlier when looking for the unique values for each columns we found that :
''' All the track names and album names are not unique this means that the row values can be repeated. '''
# This is a crucial piece when imputing values.
```

```python
# Now I am think two methods can be used for imputing : 
# 1) KNN Imputer
# 2) Mode Imputation

# Did some googling and found out that KNN and MICE is overkill , also considering the ratio of dataset size and missing row size ,
# we will use simple mode imputation.
```

```python
# Analyzing modes
df['album_name'].mode()[0] , df['track_name'].mode()[0]
```

```python
# Filling the columns with mode
df['album_name'] = df['album_name'].fillna(df['album_name'].mode()[0])
df['track_name'] = df['track_name'].fillna(df['track_name'].mode()[0])
```
