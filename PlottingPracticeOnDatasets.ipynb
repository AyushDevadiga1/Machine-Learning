```python
sns.set_palette('tab10')
```

```python
sns.pairplot(df,hue='Species')
```

```python
sns.displot(    
                data = df,
                x = 'Width',
                kde = True,
                color = 'Red'
)
```

```python
sns.displot(    
                data = df,
                x = 'Weight',
                kde = True,
                color = 'Violet'
)
```

```python
sns.countplot(
            data = df,
            x = 'Species',
            legend=True,
            palette = 'mako', # flare , inferno , magma , Paired , Accent , Reds , Blues , BuGn
            fill=True,
            hue = 'Species'
)
```
