```python
# The ravel vs flatten war 
# Both ravel and flatten are used to flatten the array but we use flatten when we want the original array being flattened to not be affected.
# We use ravel for memory effecieny where we dont care about the original array and need to using a lot of memory(when working with dataframes).
```

```python
# Lets now do this practically:
x = np.array([[1,2,3],[4,5,6]])
print(f' Original array : \n {x} \n')
print(f'Flattened array : \n {x.flatten()} \n')
x.flatten()[0] = 7 # This changes is not affected
print(f' Original array status after flattening : \n {x} \n')
```

```python
# Lets now do this practically:
x = np.array([[1,2,3],[4,5,6]])
print(f' Original array : \n {x} \n')
print(f'Flattened array using ravel : \n {x.ravel()} \n')
x.ravel()[0]=8 # This change is affected in the original array directly
print(f' Original array status after ravel : \n {x} \n')
```

```python
arr1 = np.array(
    [
        [1 , 2  , 3],
        [1 , 2  , 3],
        [1 , 2  , 3]
    ]
)
arr2 = np.array(
    [
        [1 , 0 , 0],
        [0 , 2 , 0],
        [0 , 0 , 3]
    ]
)
arr3 = np.array(
    [
        [
            [1 , 0 , 0],
            [0 , 2 , 0],
            [0 , 0 , 3]
        ]
    ]
)
arr4 = np.array(
        [1 , 0 , 0 ]
)
```

```python
# Firstly lets discuss some of the statistical functions that can be used using numpy.
print(f' The  array :\n {(arr2)}')
print(f' The mean of the array : {np.mean(arr2)}') # The average of all the elements 
print(f' The median of the array : {np.median(arr2)}') # The median / middle value when arranged in any order
print(f' The standard deviation of the array : {np.std(arr2)}') # standard deviation of the array.
print(f' The variance of the array : {np.var(arr2)}') # The variance of the array
print(f' The covariance of the array : \n {np.cov(arr2)}') # The covariance of the array
print(f' maximum element of the array : {np.max(arr2)}') # Returns the maximum element of the array 
print(f' index of maximum element of the array : {np.argmax(arr2)}') # Returns the index of maximum element of the array 
print(f' minimum element of the array : {np.min(arr2)}') # Returns the minimum element of the array 
print(f' index of minimum element of the array : {np.argmin(arr2)}') # Returns the index of maximum element of the array 
```

```python
# We can also do the dot product of matrices using the dot function
print(f'Matrix 1 : \n {(arr1)}\n ')
print(f'Matrix 2 : \n {(arr2)}\n ')
print(f'Dot product of the matrices : \n{np.dot(arr1,arr2)} \n')
# We can also use 3 matrix together like :
print(f'Dot product of 3 matrices : \n{np.dot(arr1,arr2,arr1)} \n')
```

```python
# The note for n dimensions :
'''
Input a 	Input b	    Result
1-D array	1-D array	A scalar (dot product).
1-D array	2-D array	A 1-D array (matrix-vector product).
2-D array	1-D array	A 1-D array (matrix-vector product).
2-D array	2-D array	A 2-D array (conventional matrix product).
N-D array	N-D array	N-D array (stacks of matrices in last two dims are multiplied).
'''
```

```python
# Now to calculate the inverse of the matrix we can use linalg.inv.
print(f'Original Matrix : \n {(arr2)} \n')
print(f'After Inverse : \n{np.linalg.inv(arr2)} \n')
```

```python
# We can also use the linalg.eig function to calculate the eigen values for the matrix
v,m = np.linalg.eig(arr1)
print(f'Original Matrix : \n {(arr1)} \n')
print(
    f'The eigen values are : {v} \n The eigen matrix is \n {m}'
)
# The eigen values follows a fundamental eigen value equation which is why to begin with we use them.
# The equation states that :     A.x = lambda.x
'''
    (A): The square matrix (transformation)
    (x): The eigenvector. It is a non-zero vector that does not change its direction when multiplied by the matrix
    (lambda): The eigenvalue. It is the scalar (number) that determines how much the eigenvector (x) is stretched or compressed.
'''

# Lets verify this for our matrix by using allclose function :
# All allclose checks if two arrays, a and b, are element-wise equal within a specified tolerance, returning True if all elements match according to |a - b| <= (atol + rtol * |b|), and False otherwise.
np.allclose(
    np.dot(arr1,m[:,0]),   #LHS
    np.dot(v[0],m[:,0])   #RHS
    # v[0] is the first eigenvalue, m[:, 0] is its corresponding eigenvector
)
```

```python
                                                            # Uniform distribution
# We may have this already done as the random.rand create a uniform distribution between the interval 0 and 1.
# We will reverify it with plotting the graph for it.
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.rand(1,50,100).ravel() # Ravel is used to flatten the array to 1D
)
plt.subplot(1,2,2)
sns.histplot( # Histplot for the same and I have chosen it cause if we try to use displot the graph will appear to be blank as we need axes level functions here and displot is figure levle function
    np.random.rand(1,50,100).ravel() 
)
plt.show()
# As we can see the graph is almost uniform
```

```python
                                                            # Normal distribution
# We can make a normal distribution for us by passing the mean and std with the size as params
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.normal(
        2.5 , # The mean for the normal distribution
        6 , # The std for the normal distribution
        (3,3,3) # The shape of the normal distributed matrix
    ).ravel() 
)
plt.subplot(1,2,2)
sns.histplot( 
    np.random.normal(
        2.5 , # The mean for the normal distribution
        6 , # The std for the normal distribution
        (3,3,3) # The shape of the normal distributed matrix
    ).ravel() 
)
plt.show()
# As we can see the graph is somewhat normal
```

```python
 # We can experiment more and can also make standard normal distribution as below
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.normal(
        0 , # The mean for the normal distribution
        1, # The std for the normal distribution
        (100) # The shape of the normal distributed matrix
    ).ravel(),
    fill = True,
    color="red"
)
plt.subplot(1,2,2)
sns.histplot( 
    np.random.normal(
        0 , # The mean for the normal distribution
        1 , # The std for the normal distribution
        (100) # The shape of the normal distributed matrix
    ).ravel(),
    kde=True,
    color="darkred",
    stat = 'density'
)
plt.show()
```

```python
                                                            # Log - Normal distribution
# Similarly e can make a log-normal distribution by passing the mean and std with the size as params.
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.lognormal(
        0 , 
        1 , 
        (100) # The shape of the normal distributed matrix
    ),
    color = 'darkcyan'
)
plt.subplot(1,2,2)
sns.histplot( 
    np.random.lognormal(
        0, 
        1, 
        (100) # The shape of the normal distributed matrix
    ),
    color = 'cyan'
)
# Here ravel was not necessary as I passed the rows/size only
plt.show()
# As we can see the graph is somewhat log-normal
```

```python
                                                        # Pareto Distribution
# We can also pareto distribution bys following these steps
# Pareto distribution has two parameters : a (the shape) and m(the size). 
# The formula for pareto distribution's  pdf is : p(x) = a*(m**a)/x**(a+1)

a,m = 2.5,2

s = (np.random.pareto(a,5000)+1)*m # Note : +1 is added as we need pareto 1 distribution but our pareto function gives pareto 2.
# Now lets visualize this distribution
plt.figure(figsize=(12,5))
counts , bins , _ = plt.hist(
    s,
    bins = 500,
    density = True # Scales the histogram.
)
# Now calculatin the probability for each values using formula
# Now to calculate the probability density function we will use the formula like this :
par = (a*(m**a))/(bins**(a+1)) # bins = x , rest is already known
# Now we will plot this 

plt.plot(
    bins,
    max(counts)*par/max(par),
    linewidth = 2,
    color='red'
)
plt.show()
# Plotting this graph gives results just like a pareto distribution should look like .
# A common example of this distribution is the amount of money a person has in the society.
# More people tend to have less amount of money considered with the people who have money but less in quantity.
# It mostly follows the 80-20 rule.
```

```python
                                                        # Power Distribution
# Similarly we have another distribution called powere distribution.
# Power distribution has one parameters : a (the shape)/power  
# The formula for power distribution's  pdf is : p(x) = a*x**(a-1)  x : [0,1] & a > 0 
a,sample_size = 5,1000 # The power
p = (np.random.power(a,sample_size)) 
# visualizing this distribution
plt.figure(figsize=(14,6))
counts , bins , ignored = plt.hist( # Used ignored / _ to Ignore patches that are passed 
    p,
    bins = 30
)
# First we want n values in range 0,1 because we will use those values as the input in the equation.
x = np.linspace(0,1,100) #linspace is the best for times like this 
# Now for the curve we will use the pdf equation.
y = a*x**(a-1)
# Now to calculate the y axis for the plot we will make a normal_y (scalling y ) by using : 
normed_y = sample_size*np.diff(bins)[0]*y
# Now plotting this pdf curve
plt.plot(
    x, # The x axis
    normed_y # The y axis 
)
# The distribution i power distribution and is similar to inverse pareto of the same a value.
```

```python
                                                        # Poisson Distribution
# Poisson Distribution is used to give the probabilty mass function of a random discrete variable.
# There are two parameters here in the pmf function : lambda(l) and the number of occurences(k = 0,1,2....). 
#                                                  y = (l**k)(e**-l)/k!
# Lets implement this:
pois = np.random.poisson(
    5,  # the lambda value
    1000 # the sample size
)
''' 
# This code when run will result in a bit of abnormal behaviour but u can try , as the hist returns many values ,
# we need variables to store or ignore them.
plt.hist(
    pois,
    bins = 100
)
'''
counts , bins ,ignored = plt.hist(
    pois, 
    13
)
# Below is a hist representing the poisson distribution .
```
