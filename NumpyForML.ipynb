```python
# The ravel vs flatten war 
# Both ravel and flatten are used to flatten the array but we use flatten when we want the original array being flattened to not be affected.
# We use ravel for memory effecieny where we dont care about the original array and need to using a lot of memory(when working with dataframes).
```

```python
# Lets now do this practically:
x = np.array([[1,2,3],[4,5,6]])
print(f' Original array : \n {x} \n')
print(f'Flattened array : \n {x.flatten()} \n')
x.flatten()[0] = 7 # This changes is not affected
print(f' Original array status after flattening : \n {x} \n')
```

```python
# Lets now do this practically:
x = np.array([[1,2,3],[4,5,6]])
print(f' Original array : \n {x} \n')
print(f'Flattened array using ravel : \n {x.ravel()} \n')
x.ravel()[0]=8 # This change is affected in the original array directly
print(f' Original array status after ravel : \n {x} \n')
```

```python
                                                            # Uniform distribution
# We may have this already done as the random.rand create a uniform distribution between the interval 0 and 1.
# We will reverify it with plotting the graph for it.
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.rand(1,50,100).ravel() # Ravel is used to flatten the array to 1D
)
plt.subplot(1,2,2)
sns.histplot( # Histplot for the same and I have chosen it cause if we try to use displot the graph will appear to be blank as we need axes level functions here and displot is figure levle function
    np.random.rand(1,50,100).ravel() 
)
plt.show()
# As we can see the graph is almost uniform
```

```python
                                                            # Normal distribution
# We can make a normal distribution for us by passing the mean and std with the size as params
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.normal(
        2.5 , # The mean for the normal distribution
        6 , # The std for the normal distribution
        (3,3,3) # The shape of the normal distributed matrix
    ).ravel() 
)
plt.subplot(1,2,2)
sns.histplot( 
    np.random.normal(
        2.5 , # The mean for the normal distribution
        6 , # The std for the normal distribution
        (3,3,3) # The shape of the normal distributed matrix
    ).ravel() 
)
plt.show()
# As we can see the graph is somewhat normal
```

```python
 # We can experiment more and can also make standard normal distribution as below
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.normal(
        0 , # The mean for the normal distribution
        1, # The std for the normal distribution
        (100) # The shape of the normal distributed matrix
    ).ravel(),
    fill = True,
    color="red"
)
plt.subplot(1,2,2)
sns.histplot( 
    np.random.normal(
        0 , # The mean for the normal distribution
        1 , # The std for the normal distribution
        (100) # The shape of the normal distributed matrix
    ).ravel(),
    kde=True,
    color="darkred",
    stat = 'density'
)
plt.show()
```

```python
                                                            # Log - Normal distribution
# Similarly e can make a log-normal distribution by passing the mean and std with the size as params.
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.lognormal(
        0 , 
        1 , 
        (100) # The shape of the normal distributed matrix
    ),
    color = 'darkcyan'
)
plt.subplot(1,2,2)
sns.histplot( 
    np.random.lognormal(
        0, 
        1, 
        (100) # The shape of the normal distributed matrix
    ),
    color = 'cyan'
)
# Here ravel was not necessary as I passed the rows/size only
plt.show()
# As we can see the graph is somewhat log-normal
```

```python
                                                        # Pareto Distribution
# We can also pareto distribution bys following these steps
# Pareto distribution has two parameters : a (the shape) and m(the size). 
# The formula for pareto distribution's  pdf is : p(x) = a*(m**a)/x**(a+1)

a,m = 2.5,2

s = (np.random.pareto(a,5000)+1)*m # Note : +1 is added as we need pareto 1 distribution but our pareto function gives pareto 2.
# Now lets visualize this distribution
plt.figure(figsize=(12,5))
counts , bins , _ = plt.hist(
    s,
    bins = 500,
    density = True # Scales the histogram.
)
# Now calculatin the probability for each values using formula
# Now to calculate the probability density function we will use the formula like this :
par = (a*(m**a))/(bins**(a+1)) # bins = x , rest is already known
# Now we will plot this 

plt.plot(
    bins,
    max(counts)*par/max(par),
    linewidth = 2,
    color='red'
)
plt.show()
# Plotting this graph gives results just like a pareto distribution should look like .
# A common example of this distribution is the amount of money a person has in the society.
# More people tend to have less amount of money considered with the people who have money but less in quantity.
# It mostly follows the 80-20 rule.
```

```python
                                                        # Power Distribution
# Similarly we have another distribution called powere distribution.
# Power distribution has one parameters : a (the shape)/power  
# The formula for power distribution's  pdf is : p(x) = a*x**(a-1)  x : [0,1] & a > 0 
a,sample_size = 5,1000 # The power
p = (np.random.power(a,sample_size)) 
# visualizing this distribution
plt.figure(figsize=(14,6))
counts , bins , ignored = plt.hist( # Used ignored / _ to Ignore patches that are passed 
    p,
    bins = 30
)
# First we want n values in range 0,1 because we will use those values as the input in the equation.
x = np.linspace(0,1,100) #linspace is the best for times like this 
# Now for the curve we will use the pdf equation.
y = a*x**(a-1)
# Now to calculate the y axis for the plot we will make a normal_y (scalling y ) by using : 
normed_y = sample_size*np.diff(bins)[0]*y
# Now plotting this pdf curve
plt.plot(
    x, # The x axis
    normed_y # The y axis 
)
# The distribution i power distribution and is similar to inverse pareto of the same a value.
```
