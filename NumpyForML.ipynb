```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# There is a significant difference between a normal array and numpy array .
# We can see this below as we examine their datatypes
l = [2,3,4]
n = np.array([2,3,4])
print(type(l))
print(type(n))
# As you can see that the type of the first is list and second is numpy.ndarray
# Why this ? Cause I think list is a dtype which is heterogenous in nature and we can use it to store many other datatypes other than this.
# We use numpy.ndarray specifically for numerical functions on it which can not be applied to other dtypes.
```

```python
n1 = np.random.random() # Returns a random float number between interval : 0.0 to 1.0 
# Also as we have passed no parameters the value is a float value and is of 0th dimension.
print(f'0D : {n1}\n')
n2 = np.random.random(5)
# It accepts only one parameter and when passed directly it treats it as the length for the random np.array need to be generated.
print(f'1D : {n2}\n')
n3 = np.random.random([3,3,3]) # (layer,row,column)
# If a tuple of desired shape is passed then we will have that many random values in the same interval (0.0 - 1.0)
print(f'ND : {n3}')
```

```python
np.random.random_sample([3,3]) # Another similar method is random_sample which works similarly .
```

```python
# We can use this to create scatter plots and other plots for our need whenever we need to experiment on graphs and can be also be used
# to create a dataframe of this if needed.

# Basically we can create artificial datasets and use it for our convenience when needed .

# Plotting use random class
plt.scatter(
    # scatter method is overpowered it automatically flattens this to 1D array before plotting .
    # We cant use dataframe here cause there is mixed dtypes in DataFrame and it maybe contain string,datetime and other which wont be plotted.
    # This is why for dataframes and for advance plotting we can use seaborn .
    np.random.random(50),
    np.random.random(50)**3 ,
    color = 'black'
)
```
