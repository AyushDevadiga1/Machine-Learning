```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# There is a significant difference between a normal array and numpy array .
# We can see this below as we examine their datatypes
l = [2,3,4]
n = np.array([2,3,4])
print(type(l))
print(type(n))
# As you can see that the type of the first is list and second is numpy.ndarray
# Why this ? Cause I think list is a dtype which is heterogenous in nature and we can use it to store many other datatypes other than this.
# We use numpy.ndarray specifically for numerical functions on it which can not be applied to other dtypes.
```

```python
x = np.array([[[1,2,3,4],[2,43,4,4]]])
print(f'Matrix before : {x}')
print(f'Before reshape : {x.shape}')
print(f'Matrix after : {x.reshape(2,4)}') # params (array_to_be_reshaped,new_shape)
print(f'After reshape : {x.reshape(2,4).shape}')
```

```python
# The ravel vs flatten war 
# Both ravel and flatten are used to flatten the array but we use flatten when we want the original array being flattened to not be affected.
# We use ravel for memory effecieny where we dont care about the original array and need to using a lot of memory(when working with dataframes).
```

```python
# Lets now do this practically:
x = np.array([[1,2,3],[4,5,6]])
print(f' Original array : \n {x} \n')
print(f'Flattened array : \n {x.flatten()} \n')
x.flatten()[0] = 7 # This changes is not affected
print(f' Original array status after flattening : \n {x} \n')
```

```python
# Lets now do this practically:
x = np.array([[1,2,3],[4,5,6]])
print(f' Original array : \n {x} \n')
print(f'Flattened array using ravel : \n {x.ravel()} \n')
x.ravel()[0]=8 # This change is affected in the original array directly
print(f' Original array status after ravel : \n {x} \n')
```

```python
### *** Now we will cover various distributions that can be made in numpy. *** ###
```

```python
                                                            # Uniform distribution
# We may have this already done as the random.rand create a uniform distribution between the interval 0 and 1.
# We will reverify it with plotting the graph for it.
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.rand(1,50,100).ravel() # Ravel is used to flatten the array to 1D
)
plt.subplot(1,2,2)
sns.histplot( # Histplot for the same and I have chosen it cause if we try to use displot the graph will appear to be blank as we need axes level functions here and displot is figure levle function
    np.random.rand(1,50,100).ravel() 
)
plt.show()
# As we can see the graph is almost uniform
```

```python
                                                            # Normal distribution
# We can make a normal distribution for us by passing the mean and std with the size as params
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.normal(
        2.5 , # The mean for the normal distribution
        6 , # The std for the normal distribution
        (3,3,3) # The shape of the normal distributed matrix
    ).ravel() 
)
plt.subplot(1,2,2)
sns.histplot( 
    np.random.normal(
        2.5 , # The mean for the normal distribution
        6 , # The std for the normal distribution
        (3,3,3) # The shape of the normal distributed matrix
    ).ravel() 
)
plt.show()
# As we can see the graph is somewhat normal
```

```python
 # We can experiment more and can also make standard normal distribution as below
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.kdeplot(
    np.random.normal(
        0 , # The mean for the normal distribution
        1, # The std for the normal distribution
        (100) # The shape of the normal distributed matrix
    ).ravel(),
    fill = True,
    color="royalblue"
)
plt.subplot(1,2,2)
sns.histplot( 
    np.random.normal(
        0 , # The mean for the normal distribution
        1 , # The std for the normal distribution
        (100) # The shape of the normal distributed matrix
    ).ravel(),
    kde=True,
    color="blue",
    stat = 'density'
)
plt.show()
```
