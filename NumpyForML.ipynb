```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n1 = np.random.random() # Returns a random float number between interval : 0.0 to 1.0 
# Also as we have passed no parameters the value is a float value and is of 0th dimension.
print(f'0D : {n1}\n')
n2 = np.random.random(5)
# It accepts only one parameter and when passed directly it treats it as the length for the random np.array need to be generated.
print(f'1D : {n2}\n')
n3 = np.random.random([3,3,3]) # (layer,row,column)
# If a tuple of desired shape is passed then we will have that many random values in the same interval (0.0 - 1.0)
print(f'ND : {n3}')
```
