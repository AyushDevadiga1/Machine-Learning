```python
import numpy as np
import random
from sklearn.neighbors import NearestNeighbors
```

```python
X = np.ceil(np.random.rand(10,2)*10)
```

```python
X
```

```python
nnbrs = NearestNeighbors(
    n_neighbors = 3,
    algorithm = 'ball_tree'
).fit(X)
```

```python
distances,indices = nnbrs.kneighbors(X)
```

```python
indices
```

```python
distances
```

```python
nnbrs.kneighbors_graph(X).toarray()
```

```python
min_distances = np.sort(distances[:,1:],axis=0)
```
