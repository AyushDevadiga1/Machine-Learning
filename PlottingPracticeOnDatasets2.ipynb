```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot 
```

```python
# Swarm plots can be used to plot categorical and continous numeric value 
```

```python
from sklearn.datasets import load_linnerud
```

```python
df = pd.read_csv('Titanic-dataset.csv')
```

```python
df.head()
```

```python
ax = sns.swarmplot(
    data = df,
    x = 'Pclass',
    y = 'Age',
    size = 4,
    hue = 'Sex'
)
```

```python
ax1 = sns.stripplot(
    data = df,
    x = 'Pclass',
    y = 'Age',
    size = 5,
    hue = 'Sex',
    palette='viridis' # plasma , RdBu , tab20 , Accent
)
```

```python
ax2 = sns.countplot(
    data = df,
    x = 'Survived',
    hue = 'Sex',
    palette='tab20',
    stat = 'percent'
)
```

```python
ax3 = sns.scatterplot(
    data = df,
    x = 'Age',
    y = 'Fare',
    hue = 'Embarked',
    style = 'Survived',
    palette='tab10' 
)
```
