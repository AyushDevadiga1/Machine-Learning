```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Swarm plots can be used to plot categorical and continous numeric value 
```

```python
from sklearn.datasets import load_linnerud
```

```python
df = pd.read_csv('Titanic-dataset.csv')
```

```python
df.head()
```

```python
ax = sns.swarmplot(
    data = df,
    x = 'Pclass',
    y = 'Age',
    size = 4,
    hue = 'Sex'
)
```

```python
ax1 = sns.stripplot(
    data = df,
    x = 'Pclass',
    y = 'Age',
    size = 5,
    hue = 'Sex',
    palette='viridis' # plasma , RdBu , tab20 , Accent
)
```

```python
ax2 = sns.countplot(
    data = df,
    x = 'Survived',
    hue = 'Sex',
    palette='tab20',
    stat = 'percent'
)
```

```python
# sns.set_theme('notebook') # Set theme
# sns.set_style('ticks') # Set style
plt.figure(figsize=(12,6))
ax3 = sns.scatterplot(
    data = df,
    x = 'Age',
    y = 'Fare',
    hue = 'Embarked',
    style = 'Survived',
    palette='tab10' 
)
# plt.xlim(0,100)  #xlimits
plt.ylim(0,600)  #ylimits
plt.show()
```

```python
# We use histplot to visualize continous variables 
plt.figure(figsize=(13,6))
sns.histplot(
    data = df ,
    x = 'Age',
    # binrange=(0,100),
    # bins = 25,
    legend=True,
    hue = 'Sex',
    thresh = 1,
    kde = True,
    multiple = 'stack',
    palette = 'tab10',
    # element = 'bars'
)
```

```python
# Kdeplot is similar to histplot and is used to estimate the density of the variable 
plt.figure(figsize=(12,5))
sns.kdeplot(
    data = df ,
    x = 'Age',
    legend=True,
    hue = 'Pclass',
    fill = True,
    alpha = 0.35,
    palette = 'tab10',
)
```

```python
# Jointplot used to use two analyze variables together using two techniques consequently 
sns.jointplot(
    data = df,
    x = 'Age',
    y = 'Fare',
    kind = 'reg', # 'scatter', 'kde', 'hist', 'hex', 'reg', 'resid'
)
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
# We use joint plots to get a overall view of the dataset using any univarate plotting method 
sns.pairplot(
    data = data[['Glucose' , 'BloodPressure' , 'BMI' , 'Age' , 'Outcome']], # Use with caution as loading many columns together takes time
    kind = 'reg', # We can use many other types : ['scatter', 'kde', 'hist', 'reg']
    hue = 'Outcome'
)
```

```python
# Works with other plots to make comprehensive analysis.
sns.rugplot(
    data = data,
    x = 'Insulin',
    height = 0.5,
    hue = 'Outcome',
    palette = ['yellow' , 'blue'],
)
```
