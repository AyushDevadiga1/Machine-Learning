```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df
```

### The first steps when loading any data must be to check for :
    1. Shape - to know the amount of rows and columns to expect; also the size of data .
    2. Sample - to look for rows in random places to make sure further analysis are not delusional.
    3. DataTypes - to check for the data types of each columns via info().
    4. Errors - to check for null / duplicates / NaN values and preprocess accordingly.
    5. Description -  to get a statistical analysis of all the columns in data.
    6. Correlation - to check for the relationship between different variables and then proceed accordingly.

Analysing Categorical Data

```python
# As we have categorical data we will find countplot
sns.countplot(x = df['Survived'], palette='Set2')
```

```python
# sns.countplot(x=df['Pclass'] ,palette = 'Set1') # Old method 
sns.countplot(x='Pclass', data = df , hue = 'Pclass' , palette = 'Set1' , legend = False)
```

```python
sns.countplot(x = df['Embarked'])
```

```python
# The plot can be used to plot many charts , one of them is piechart .
# Make sure we are using the value_counts() function cause else dataframe will be returned and will result in error .
df['Survived'].value_counts().plot(kind = 'pie' , autopct = '% .3f')
```

Analyzing Numerical Data

```python
# We use histogram to check the distribution followed here and the skewness.
plt.hist(df['Age'],bins=15)
plt.plot()
```

```python
# Displot similar to histogram but also has PDF through which we can predict the probality of the person with his age to check whether he survived or not .
sns.displot(df['Age'] , kde=True)
```

```python
# The box-plot is used to find the outliers,the min,max,range,etc.
sns.boxplot( x = df['Fare'] , width=0.9 )
```

```python
sns.boxplot( x = df['Age'] )
```
